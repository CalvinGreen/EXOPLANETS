```python
import numpy as np
import pandas as pd 
import keras
import sklearn 
from sklearn.cross_validation import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve, roc_auc_score
from scipy.ndimage.filters import uniform_filter1d

print('IMPORTING THE EXOPLANET TRAINING DATASET')

raw_data = np.loadtxt('exoTrain.csv', skiprows=1, delimiter=',')

x_train,y_train =  raw_data[:, 1:], raw_data[:, 0, np.newaxis] - 1.

raw_data = np.loadtxt('exoTest.csv', skiprows=1, delimiter=',')
print('IMPORTING THE EXOPLANET TESTING DATASET')
x_test,y_test  = raw_data[:, 1:],raw_data[:, 0, np.newaxis] - 1.

del raw_data
print('x test shape',x_test.shape[0])
print('y test shape',y_test.shape[0])
x_train,x_test = x_train[2087:], x_test[:3000] #.4 TEST SIZE

y_train,y_test = y_train[2087:], y_test[:3000]
print('X TEST')
display(x_test[:5])
print('Y TEST')
display(y_test[:5])
print('X TRAIN')
display(x_train[:5])
print('Y TRAIN')
display(y_train[:5])
print(x_train.shape[0])
print(y_train.shape[0])
```

```python
print('CREATING R2 PERFOMANCE METRIC')
def performance_metric(y_true, y_predict):
    from sklearn.metrics import r2_score,roc_curve, roc_auc_score
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)

    # Return the score
    return score
print('CREATED PERFOMANCE METRIC')
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import ShuffleSplit
print('fitting the data to an ADABOOSTREGRESSOR \n , params:= \n n_estimators:[200,150,100,50,25,20,10,5], \n learning_rate:[.7,.6,.5,.4,.25,.1], \n LOSS: linear, square,exponential')
print('fitting the ADABOOSTREGRESSOR to a DECISIONTREEREGRESSOR \n \n \n  ')
def fit_model(x, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    cv_sets = ShuffleSplit(x.shape[0], n_iter = 15, test_size = 0.20, random_state =None)
    
    # TODO: Create a decision tree regressor object
    regressor = AdaBoostRegressor(DecisionTreeRegressor(criterion='mae',max_depth=300,max_leaf_nodes=165, min_samples_leaf=12, min_samples_split=9))

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'n_estimators':[200,150,100,50,25,20,10,5],
          'learning_rate':[.7,.6,.5,.4,.25,.1],
          'loss':['linear','square','exponential'],
             }

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # TODO: Create the grid search object
    grid = GridSearchCV(regressor, params, scoring = scoring_fnc, cv = cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(x, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

```python
print('COFFE BREAK \n FITTING THE GRID TO THE PARAMETERS.. \n NAPS ARE NICE TOO ^__^')
reg = fit_model(x_train, y_train.ravel())
print('BEEP!! BEEP!!! BEEEEEPPP!!!')
```

```python
from sklearn.decomposition import PCA
PCA = PCA(n_components= 10, svd_solver='randomized', whiten=True, tol=0.0025, iterated_power=500)
print(PCA)

print('FITTING/transforming the ica,pca data')

from sklearn.decomposition import FastICA
ICA1= FastICA(n_components= 10, algorithm='deflation', whiten=True, fun='logcosh', fun_args= {'alpha' :1.5}, max_iter=500, tol=0.0025, random_state=11111)
print(ICA1)
ICA1.fit(x_train)
print('ICA')

ICA_T= ICA1.fit_transform(x_train)

print('STARTING PCA')
PCA_T = PCA.fit_transform(x_train)

print('finished')

```

```python
print('CREATING VISUAL')

models = [ICA1.mixing_, x_train, ICA_T, PCA_T]
names = ['Observations (mixed signal)',
         'True Sources',
         'ICA recovered signals',
         'PCA recovered signals']
colors = ['red', 'steelblue', 'orange','purple',
         'green']

for ii, (model, name) in enumerate(zip(models, names), 1):
    plt.subplot(4, 1, ii)
    plt.title(name)
    plt.legend(models)
    for sig, color in zip(model.T, colors):
        plt.plot(sig, color=color)


# plt.subplots_adjust(0.09, 0.04, 0.94, 0.94, 0.26, 0.46)
plt.show()
```
